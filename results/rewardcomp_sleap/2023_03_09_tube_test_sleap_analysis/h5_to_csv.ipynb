```python
from scipy.interpolate import interp1d

def fill_missing(Y, kind="linear"):
    """Fills missing values independently along each dimension after the first."""

    # Store initial shape.
    initial_shape = Y.shape

    # Flatten after first dim.
    Y = Y.reshape((initial_shape[0], -1))

    # Interpolate along each slice.
    for i in range(Y.shape[-1]):
        y = Y[:, i]

        # Build interpolant.
        x = np.flatnonzero(~np.isnan(y))
        f = interp1d(x, y[x], kind=kind, fill_value=np.nan, bounds_error=False)

        # Fill missing
        xq = np.flatnonzero(np.isnan(y))
        y[xq] = f(xq)
        
        # Fill leading or trailing NaNs with the nearest non-NaN values
        mask = np.isnan(y)
        y[mask] = np.interp(np.flatnonzero(mask), np.flatnonzero(~mask), y[~mask])

        # Save slice
        Y[:, i] = y

    # Restore to initial shape.
    Y = Y.reshape(initial_shape)

    return Y
```

```python
def produce_df(rows, columns, row_names=None, column_names=None):
    """rows is a list of lists that will be used to build a MultiIndex
    columns is a list of lists that will be used to build a MultiIndex"""
    row_index = pd.MultiIndex.from_product(rows, names=row_names)
    col_index = pd.MultiIndex.from_product(columns, names=column_names)
    return pd.DataFrame(index=row_index, columns=col_index)
```

# Getting the coordinates of one animal

- Flattening (3D of frame, node, x or y to 2D of frame, node/x or y)the list so that it can be used in a dataframe
    - https://www.delftstack.com/howto/numpy/convert-3d-array-to-2d-array-in-python/

```python
def convert_single_subj_arr_from_3d_to_2d(single_animal_array):
    """
    """
    return single_animal_array.reshape(single_animal_array.shape[0], (single_animal_array.shape[1]*single_animal_array.shape[2]))
```

```python
first_animal_2D_arr = convert_single_subj_arr_from_3d_to_2d(locations[:,:,:,0])
```

- Creating headers for animal one
    - https://stackoverflow.com/questions/32370402/giving-a-column-multiple-indexes-headers

```python
first_animal_header = pd.MultiIndex.from_product([["first_animal"], node_names,
                                     ['x', 'y']],
                                    names=["animal_id", 'body_part','x_or_y'])
```

```python
first_animal_df = pd.DataFrame(first_animal_2D_arr, columns=first_animal_header)
```

```python
first_animal_df
```

# Combining with the second animal's array

```python
second_animal_2D_arr = convert_single_subj_arr_from_3d_to_2d(locations[:,:,:,1])
```

```python
second_animal_header = pd.MultiIndex.from_product([["second_animal"], node_names,
                                     ['x', 'y']],
                                    names=["animal_id", 'body_part','x_or_y'])
```

```python
second_animal_df = pd.DataFrame(second_animal_2D_arr, columns=second_animal_header)
```

```python
second_animal_df
```

- Combining the dataframes for the first and second animals

```python
first_and_second_animal_df = pd.merge(first_animal_df, second_animal_df, left_index=True, right_index=True)
```

```python
video_to_everything = defaultdict(dict)
pose_output_dir = "./proc/tube_test-Apr16-2023/videos/pose_estimation"

for h5_filepath in glob.glob("./data/cd1_predicted_frames/*h5"):
    with h5py.File(h5_filepath, "r") as f:
        locations = f["tracks"][:].T
        tracking_scores = f["tracking_scores"][:]
        instances_scores = f["instance_scores"][:]
        locations = fill_missing(locations)
    first_animal_2D_arr = convert_single_subj_arr_from_3d_to_2d(locations[:,:,:,0])
    first_animal_df = pd.DataFrame(first_animal_2D_arr, columns=first_animal_header)
    
    second_animal_2D_arr = convert_single_subj_arr_from_3d_to_2d(locations[:,:,:,1])
    second_animal_df = pd.DataFrame(second_animal_2D_arr, columns=second_animal_header)
    
    first_and_second_animal_df = pd.merge(first_animal_df, second_animal_df, left_index=True, right_index=True)
    
    h5_filename = os.path.basename(h5_filepath)
    h5_csv_name = ".".join([h5_filename.split(".")[0], "csv"])
    
    first_animal_df.to_csv(os.path.join(pose_output_dir, h5_csv_name))  
#     first_and_second_animal_df.to_csv(os.path.join(pose_output_dir, h5_csv_name))
```
